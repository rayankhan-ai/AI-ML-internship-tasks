```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Load a Dataset using pandas")
dataset = pd.read_csv(r'C:\Users\ahmad\Desktop\ML  Dataset\Iris.csv')
```

```python
print("Number of Rows And Columns of a Dataset")
dataset.shape
```

```python
print("Columns Names")
dataset.columns
```

```python
print("First few Rows of a Dataset")
dataset.head()
```

```python
print("Information About a Dataset")
dataset.info()
# Total number of rows and columns
# Column names and their index number
# Non-Null Count → how many values are not missing
# Dtype → data type of each column
# Memory usage → how much RAM dataset is using
```

```python
dataset.drop(columns=["Id"],inplace=True)
```

```python
print("Statistic Summary of Dataset")
dataset.describe()
# count : Total number of non-null values per column
# mean  : Average value of each feature
# std   : How spread out values are from mean
# min   : Smallest value in each feature
# 25%   : First quartile (Q1)
# 50%   : Second quartile (Median)
# 75%   : Third quartile (Q3)
# max   : Largest value in each feature
```

```python
# PLOT 1: SEPAL FEATURES
# x axis : SepalLengthCm (Horizontal)
# y axis : SepalWidthCm  (Vertical)
# hue    : Species (Color differentiates 3 species)
# Purpose  : To observe how Sepal Length and Sepal Width
#            relate to each other across different species
print("Relationship Between Features")
sns.scatterplot(data=dataset,x="SepalLengthCm",y="SepalWidthCm",hue="Species")
plt.title("Relationship Between SepalLengthCm and SepalWidthCm")
plt.show()
# Result: Setosa is clearly separated in top-left region
# Versicolor and Virginica overlap each other

# PLOT 2: PETAL FEATURES
# x axis : PetalLengthCm (Horizontal)
# y axis : PetalWidthCm  (Vertical)
# hue    : Species (Color differentiates 3 species)
# Purpose  : To observe how Petal Length and Petal Width
#            relate to each other across different species
sns.scatterplot(data=dataset,x="PetalLengthCm",y="PetalWidthCm",hue="Species")
plt.title("Relationship Between PetalLengthCm and PetalWidthCm")
plt.show()
# Result: All 3 species are clearly separated
# Petal features are stronger for classification
```

```python
# PLOT 1: SEPAL LENGTH DISTRIBUTION
# x axis   : SepalLengthCm (Feature values)
# y axis   : Count (Frequency of values)
# hue      : Species (Color differentiates 3 species)
# kde=True : Adds smooth density curve on top of bars
# figsize  : (7,5) sets width=7 and height=5 inches
# Purpose: To see how Sepal Length values are
# distributed across all 3 species
plt.figure(figsize=(7,5))
print("Histograms to show Value Distribuition")
sns.histplot(data=dataset,x="SepalLengthCm",hue="Species",kde=True)
plt.title("SepalLengthCm")
plt.show()
# Result: Values range mostly between 4.5 to 7.5 cm
# Setosa has shorter Sepal Length
# Virginica has longer Sepal Length

# PLOT 2: SEPAL WIDTH DISTRIBUTION
# x axis   : SepalWidthCm (Feature values)
# y axis   : Count (Frequency of values)
# hue      : Species (Color differentiates 3 species)
# kde=True : Adds smooth density curve on top of bars
# Purpose    : To see how Sepal Width values are
#              distributed across all 3 species
plt.figure(figsize=(7,5))
sns.histplot(data=dataset,x="SepalWidthCm",hue="Species",kde=True)
plt.title("SepalWidthCm")
plt.show()
# Result: Most normally distributed feature
# Setosa has widest Sepal Width
# Values concentrated between 2.5 to 3.5 cm

# PLOT 3: PETAL LENGTH DISTRIBUTION
# x axis   : PetalLengthCm (Feature values)
# y axis   : Count (Frequency of values)
# hue      : Species (Color differentiates 3 species)
# kde=True : Adds smooth density curve on top of bars
# Purpose: To see how Petal Length values are
# distributed across all 3 species
plt.figure(figsize=(7,5))
sns.histplot(data=dataset,x="PetalLengthCm",hue="Species",kde=True)
plt.title("PetalLengthCm")
plt.show()
# Result: Shows TWO peaks (Bimodal Distribution)
# Setosa peak at 1-2 cm (very small petals)
# Versicolor & Virginica peak at 4-6 cm

# PLOT 4: PETAL WIDTH DISTRIBUTION
# x axis   : PetalWidthCm (Feature values)
# y axis   : Count (Frequency of values)
# hue      : Species (Color differentiates 3 species)
# kde=True : Adds smooth density curve on top of bars
# Purpose: To see how Petal Width values are
# distributed across all 3 species
plt.figure(figsize=(7,5))
sns.histplot(data=dataset,x="PetalWidthCm",hue="Species",kde=True)
plt.title("PetalWidthCm")
plt.show()
# Result: Shows TWO peaks (Bimodal Distribution)
# Setosa tightly packed near 0.1 to 0.5 cm
#              Virginica spread between 1.5 to 2.5 cm
```

```python
print("Identifying Outliers Per Species in Each Feature")
# PLOT 1: SEPAL LENGTH BOX PLOT PER SPECIES
# x axis  : Species (3 categories on horizontal axis)
# y axis  : SepalLengthCm (Feature values on vertical)
# Purpose   : To compare Sepal Length distribution and
# identify outliers in each species separately
sns.boxplot(data=dataset,x="Species",y="SepalLengthCm")
plt.title("SepalLengthCm")
plt.show()
# Result:
# Setosa     → Shortest Sepal Length (4.3 - 5.8 cm)
# Versicolor → Medium  Sepal Length (4.9 - 7.0 cm)
# Virginica  → Longest Sepal Length (5.5 - 7.9 cm)
# Outliers   → Few outliers visible in Setosa

# PLOT 2: SEPAL WIDTH BOX PLOT PER SPECIES
# x axis  : Species (3 categories on horizontal axis)
# y axis  : SepalWidthCm (Feature values on vertical)
# Purpose   : To compare Sepal Width distribution and
# identify outliers in each species separately
sns.boxplot(data=dataset,x="Species",y="SepalWidthCm")
plt.title("SepalWidthCm")
plt.show()
# Result:
# Setosa     → Widest Sepal Width  (2.3 - 4.4 cm)
# Versicolor → Narrow Sepal Width  (2.0 - 3.4 cm)
# Virginica  → Moderate Sepal Width(2.2 - 3.8 cm)
# Outliers   → Most outliers visible in this feature

# PLOT 3: PETAL LENGTH BOX PLOT PER SPECIES
# x axis  : Species (3 categories on horizontal axis)
# y axis  : PetalLengthCm (Feature values on vertical)
# Purpose   : To compare Petal Length distribution and
# identify outliers in each species separately
sns.boxplot(data=dataset,x="Species",y="PetalLengthCm")
plt.title("PetalLengthCm")
plt.show()
# Result:
# Setosa     → Smallest Petal Length (1.0 - 1.9 cm)
# Versicolor → Medium  Petal Length  (3.0 - 5.1 cm)
# Virginica  → Largest Petal Length  (4.5 - 6.9 cm)
# Outliers   → Minimal outliers, very clean data

# PLOT 4: PETAL WIDTH BOX PLOT PER SPECIES
# x axis  : Species (3 categories on horizontal axis)
# y axis  : PetalWidthCm (Feature values on vertical)
# Purpose   : To compare Petal Width distribution and
# identify outliers in each species separately
sns.boxplot(data=dataset,x="Species",y="PetalWidthCm")
plt.title("PetalWidthCm")
plt.show()
# Result:
# Setosa     → Smallest Petal Width (0.1 - 0.6 cm)
# Versicolor → Medium  Petal Width  (1.0 - 1.8 cm)
# Virginica  → Largest Petal Width  (1.4 - 2.5 cm)
# Outliers   → Few outliers in Virginica
```
